```python
India["STATE"] = India["STATE"].str.strip()

state = India[India["STATE"] == "ANDHRA PRADESH"]

ax = India.plot(color="gray", edgecolor="black", figsize=(10,10))

state.plot(ax=ax, color="red")
plt.title("Highlighting Andhra Pradesh", fontsize=15)

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

India["STATE"] = India["STATE"].str.strip()

state1 = India[India["STATE"] == "ANDHRA PRADESH"]
state2 = India[India["STATE"] == "ASSAM"]

ax = India.plot(color="lightgray", edgecolor="black", figsize=(8,6))
state1.plot(ax=ax, color="red")
state2.plot(ax=ax, color="red")
plt.title("Highlighting Andhra Pradesh and Assam", fontsize=15)
plt.show()
```

```python
south_states = India[India["STATE"].isin(["KARNATAKA", "KERALA", "TAMIL NADU", "ANDHRA PRADESH"])]
south_states.plot(cmap="coolwarm", edgecolor="black", figsize=(8,6))
plt.title("South Indian States", fontsize=20)
plt.show()
```

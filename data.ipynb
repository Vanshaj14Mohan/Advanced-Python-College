```python
#Move dot without forming a line, move only dot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animat
from IPython.display import HTML

plt.ioff()

x = [1,2,3]
y = [1,2,3]

fig =plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

ax.set_xlim(0,4)
ax.set_ylim(0,4)

point, =ax.plot([], [], "bo")

def fun(frame):
    #ax.plot(x,y)
    point.set_data([x[frame]], [y[frame]])
    return point

a = animat.FuncAnimation(fig, fun, frames= np.arange(0,3), interval=200)

HTML(a.to_jshtml())
```

```python
#Move dot forming a line animation
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

x, y = [1, 2, 3], [1, 2, 3]
fig, ax = plt.subplots(figsize=(4, 4))
ax.set(xlim=(0, 4), ylim=(0, 4))

dot, = ax.plot([], [], "bo")
line, = ax.plot([], [], "b-")
trail = True

def fun(frame):
    dot.set_data(x[frame], y[frame])
    if trail: line.set_data(x[:frame+1], y[:frame+1])
    return (dot, line) if trail else (dot,)

a = animation.FuncAnimation(fig, fun, frames=len(x), interval=500)
plt.close(fig)
HTML(a.to_jshtml())
```

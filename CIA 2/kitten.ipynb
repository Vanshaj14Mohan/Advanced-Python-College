```python
#import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel(r'E:\Advanced Python\CIA 2\big_kitten_dataset.xlsx')
data.head(5)
```

```python
#1 Line graph of Age_Months vs Weight_kg
plt.plot(data['Age_Months'], data['Weight_kg'], marker='x')
plt.xlabel('Age (Months)')
plt.ylabel('Weight (kg)')
plt.title('Age vs Weight')
plt.grid(False)
plt.show()
```

```python
#2 Pie chart of Favorite_Toy distribution
plt.figure(figsize=(8, 8))
data['Favorite_Toy'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Favorite Toy Distribution')
plt.ylabel('')
plt.show()
```

```python
#3 Bar Graph Breed bs Average Energy Level
plt.bar(data['Breed'], data['Energy_Level'])
plt.xlabel('Breed')
plt.ylabel('Energy_Level')
plt.title('Energy_Level by Breed')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("E:\Advanced Python\Assignment 2\Sales.xlsx")
data.head()
data.tail()
```

```python
#1.  Plot a line graph showing the Total Sales over the years. Label the x-axis as "Year" and the y-axis as "Total Sales.
plt.plot(data["Total Sales"], data["Year"])
plt.xlabel("Year")
plt.ylabel("Total Sales")
```

```python
#2-> Draw a line graph with two lines, one for Total Sales and another for Profit over the years. Use different colors and a legend.
plt.plot(data["Total Sales"], data["Year"],label="Total Sales")
plt.plot(data["Profit"], data["Year"], label="Profit")
plt.legend()
plt.show()
# plt.xlabel("Year")
# plt.ylabel("Total Sales")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")

ax.plot(data["Total Sales"], data["Year"], color = "green",linestyle="dashed")
ax.plot(data["Profit"], data["Year"], color="Yellow",linestyle="dashdot")

#ax.set_yticks([1,8])  #Setting subject positions
ax.set_yticklabels(["Year","Total Sales"])
ax.set_title("3D Sales Graph")

ax.legend()
plt.show()
```

```python
#3 ->Plot a 3D line graph showing the trend of Q1, Q2, Q3, and Q4 sales over the years in a single 3D plot. Each quarter should have a different colored line.
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

quarters = ["Q1 Sales", "Q2 Sales", "Q3 Sales", "Q4 Sales"]
colors = ["blue", "green", "red", "purple"]

for i, quarter in enumerate(quarters):
    ax.plot(data["Year"], data[quarter], zs=i, zdir='y', color=colors[i], label=quarter)

ax.set_title("Quarterly Sales Trends Over the Years (3D Line Plot)")
ax.set_xlabel("Year")
ax.set_ylabel("Quarter")
ax.set_yticks(range(4))
ax.set_yticklabels(quarters)
ax.set_zlabel("Sales")
ax.legend()
plt.show()
```

```python
#9. Visualize all four quarterly sales (Q1, Q2, Q3, Q4) in a stacked bar chart where each year has stacked bars representing the four quarters.
plt.figure(figsize=(10,8))
plt.bar(data['Year'],data['Q1 Sales'])
plt.bar(data['Year'],data['Q2 Sales'], bottom=data['Q1 Sales'])
plt.bar(data['Year'], data['Q3 Sales'],bottom=data['Q1 Sales']+data['Q2 Sales'])
plt.bar(data['Year'], data['Q4 Sales'],bottom=data['Q1 Sales']+data['Q2 Sales']+data['Q3 Sales'])


plt.xticks(rotation=90)

plt.xlabel("Year")
plt.ylabel("Quarterly Sales")
plt.title("Stacked 2d bar graph")
```

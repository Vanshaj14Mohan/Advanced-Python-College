Image Enhancer - Brightness, Sharpness, Color and Contrast

```python
from PIL import ImageEnhance
```

```python
image_col = ImageEnhance.Color(image)
color_img = image_col.enhance(1.5)
color_img.show()
```

```python
image_con = ImageEnhance.Contrast(image)
contrast_img = image_con.enhance(4)
contrast_img.show()
```

```python
image_sharp = ImageEnhance.Sharpness(image)
sharp_img = image_sharp.enhance(10)
sharp_img.show()
```

```python
image_bright = ImageEnhance.Brightness(image)
bright_img = image_bright.enhance(1)
bright_img.show()
```

```python
from PIL import ImageFilter
```

```python
#1: 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

plt.ioff()
fig, ax = plt.subplots()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

circle = plt.Circle((0, 0), 1, fill=False, edgecolor='blue')
ax.add_patch(circle)

def animate(frame):
    radius = 1 + 0.1 * frame
    circle.set_radius(radius)
    return circle,

ani = animation.FuncAnimation(fig, animate, frames=50, interval=50, blit=True)
HTML(ani.to_jshtml())
```

```python
#2. Implement a dot animation which move from rigth to left diagonal
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML

plt.ioff()
fig, ax = plt.subplots()
ax.set_xlim(0, 200) 
ax.set_ylim(0, 20)  

x = np.linspace(200, 0, 150) 
y = np.linspace(20, 0, 150) 

point, = ax.plot([], [], 'o', markersize=15, color='red')

def funcani(frame):
    point.set_data([x[frame]], [y[frame]]) 
    return point,

a = animation.FuncAnimation(fig, funcani, frames=np.arange(0, 150, 1), interval=30)

HTML(a.to_jshtml())
```

```python
#3. 
import matplotlib.pyplot as plt
plt.ioff()

fig, ax = plt.subplots(figsize=(8, 8)) 
ax.set_xlim(-3, 3) 
ax.set_ylim(-3, 3)

# Bottom
plt.plot([-1.5, 1.5], [-1.5, -1.5], 'k-', linewidth=4)  
# Left wall
plt.plot([-1.5, -1.5], [-1.5, 1], 'k-', linewidth=4) 
# Right wall
plt.plot([1.5, 1.5], [-1.5, 1], 'k-', linewidth=4)  
# Top of house
plt.plot([-1.5, 1.5], [1, 1], 'k-', linewidth=4)  
# Roof
plt.plot([-1.5, 0, 1.5], [1, 2.5, 1], 'k-', linewidth=4) 
# Door
plt.plot([-0.5, -0.5, 0.5, 0.5, -0.5], [-1.5, -0.5, -0.5, -1.5, -1.5], 'pink', linewidth=4) 

plt.show()
```

```python
#4. Exclude Rajasthan and Gujarat and map it
filtered_states = India[~India["STATE"].isin(["RAJASTHAN", "GUJARAT"])]

# Plot the filtered states
filtered_states.plot(cmap="plasma", edgecolor="black", figsize=(10,8))
plt.title("Indian States (Excluding Rajasthan and Gujarat)", fontsize=20)
plt.show()
```

```python
#5: Draw Only western Indian states in this RAJASTHAN, GUJARAT, MAHARASHTRA, GOA
west_states = India[India["STATE"].isin(["RAJASTHAN", "GUJARAT", "MAHARASHTRA", "GOA"])]
west_states.plot(cmap="plasma", edgecolor="black", figsize=(10,8))
plt.title("West Indian States", fontsize=20)
plt.show()
```

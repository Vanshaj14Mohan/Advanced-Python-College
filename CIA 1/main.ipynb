```python
#6: Generate a 3D scatter plot with Temperature, Humidity adn AQI.
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data from the Excel file
data = pd.read_excel('Weather.xlsx')

# Extract the required columns
temperature = data["Temperature (°C)"]
humidity = data["Humidity (%)"]
air = data["Air Quality Index (AQI)"]
cities = data["City"]

# Create the 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(temperature, humidity, air, c=air, cmap='coolwarm', s=100, edgecolors='black')

# Set labels and title
ax.set_xlabel("Temperature (°C)")
ax.set_ylabel("Humidity (%)")
ax.set_zlabel("Air Quality Index (AQI)")
ax.set_title("3D Scatter Plot: Temperature vs Humidity vs AQI")

# Annotate the points with city names
for i, city in enumerate(cities):
    ax.text(temperature[i], humidity[i], air[i], city, fontsize=9)

# Add a color bar
plt.colorbar(sc, ax=ax, label="AQI Level")

# Show the plot
plt.show()
```

```python
#7: Generate a 3 different lines for temperature, humidity, AQI across different geographical points.
temperature = data["Temperature (°C)"]
humdidity = data["Humidity (%)"]
air = data["Air Quality Index (AQI)"]

plt.plot(temperature, label="Temperature", marker="o")
plt.plot(humidity, label="Humidity", marker="+")
plt.plot(air, label="Air Quality Index", marker="*")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.legend()
plt.show()

```

```python
#8: 3D Line Plot – Revenue vs. Units Sold Across Regions
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

revenue = df["Revenue ($)"]
units_sold = df["Units Sold"]
regions = df["Region"]

# Convert regions to numeric indices for plotting
region_indices = range(len(regions))  # Convert categorical region names to numeric indices

# Create a 3D line plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot(region_indices, units_sold, revenue, marker='o', linestyle='-', color='blue', label="Revenue vs. Units Sold")

# Label axes
ax.set_xlabel("Region")
ax.set_ylabel("Units Sold")
ax.set_zlabel("Revenue ($)")
ax.set_title("3D Line Plot – Revenue vs. Units Sold Across Regions")

ax.set_xticks(region_indices)
ax.set_xticklabels(regions, rotation=45)  

plt.legend()
plt.show()
```

```python
#9: 3D Bar Chart – Comparing Units Sold, Revenue, and Profit Margin
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

regions = df["Region"]
units_sold = df["Units Sold"]
revenue = df["Revenue ($)"]
profit_margin = df["Profit Margin (%)"]

# Convert regions to numeric indices for plotting
region_indices = np.arange(len(regions))  # Numeric indices for categorical regions
bar_width = 0.3  # Width of bars

# Create a 3D bar chart
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot bars for each metric
ax.bar(region_indices, units_sold, zs=0, zdir='y', width=bar_width, color='b', alpha=0.7, label="Units Sold")
ax.bar(region_indices + bar_width, revenue, zs=0, zdir='y', width=bar_width, color='g', alpha=0.7, label="Revenue ($)")
ax.bar(region_indices + 2 * bar_width, profit_margin, zs=0, zdir='y', width=bar_width, color='r', alpha=0.7, label="Profit Margin (%)")

# Label axes
ax.set_xlabel("Region")
ax.set_ylabel("Category")
ax.set_zlabel("Values")
ax.set_title("3D Bar Chart – Comparing Units Sold, Revenue, and Profit Margin")

# Set custom X-axis labels for regions
ax.set_xticks(region_indices + bar_width)
ax.set_xticklabels(regions, rotation=45)  # Show region names instead of numbers

# Show legend and plot
plt.legend()
plt.show()
```

```python
#10: 3D Scatter Plot – Profit Margin vs Revenue vs Customer Rating
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

profit_margin = df["Profit Margin (%)"]  # X-axis
revenue = df["Revenue ($)"]  # Y-axis
customer_rating = df["Customer Rating"]  # Z-axis
regions = df["Region"]  # For labeling points

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

#Scatter Plot
sc = ax.scatter(profit_margin, revenue, customer_rating, c=customer_rating, cmap='coolwarm', s=100, edgecolors='black')

ax.set_xlabel("Profit Margin (%)")
ax.set_ylabel("Revenue ($)")
ax.set_zlabel("Customer Rating")
ax.set_title("3D Scatter Plot – Profit Margin vs Revenue vs Customer Rating")

# Add region labels to points
for i, region in enumerate(regions):
    ax.text(profit_margin[i], revenue[i], customer_rating[i], region, fontsize=9)

plt.colorbar(sc, ax=ax, label="Customer Rating")

plt.show()
```
